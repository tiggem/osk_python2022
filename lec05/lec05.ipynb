# Matplotlibの基本事項
- まずは、Matplotlibの可視化を体験

```python
# グラフのサイズの指定
plt.figure(figsize=(4,4)) #横4cm,縦4cm?

# 引数に指定された値のプロット。引数として、描画する点のx座標, y座標の設定、凡例に表示する文字列の設定
plt.plot([0,2,4,6,8], [0,6,12,24,48],  label="graph") 
#plotは折れ線グラフを描画する関数
#xの値 [0,2, 4, 6, 8]
#yの値 [0,6,12,24,48]

# x, y軸のラベルの設定
plt.xlabel("x axis")
plt.ylabel("y axis")

# タイトルの設定
plt.title("title")

#　凡例の表示
plt.legend()

# グラフの出力
plt.show()
```

- 次の三つの曲線をプロット

$$
y = x^2 \\
z = (x - 1)^2 \\
w = {(x - 2)(x - 4)}^2 \\
$$

```python
# 0から10まで、20個の等差数列を生成
x=np.linspace(-3,10,20)
x
```

```python
y=x**2
z=(x-1)**3
w=(x-2)*((x-4)**2)
# 横軸にx, 縦軸にそれぞれy, z, wを指定し、それぞれを判別できるラベルを凡例とする
plt.plot(x,y,label="y=x^2")
plt.plot(x,z,label="y=(x-1)^3")
plt.plot(x,w,label="y=((x-2)(x-4))^2")
# 凡例の表示
plt.legend()
```

# 複数のグラフを別々に描く

```python
plt.figure(figsize = (12,4)) #全体 横12cm 縦4cm

# plt.subplotを用いて 行x列数分だけグラフを分割して表示
# subplot(行, 列, 表示する位置)と引数を与える

# 1列目に表示
plt.subplot(1,3,1) #(1,3,1)1行3列の1つ目
plt.plot(x,y,label="y=x^2")
plt.legend()

# 2列目に表示
plt.subplot(1,3,2)
plt.plot(x,z,label="y=(x-1)^3")
plt.legend()

# 3列目に表示
plt.subplot(1,3,3)
plt.plot(x,w,label="y=((x-2)(x-4))^2")
plt.legend()

#グラフ同士の重なりを防ぐ
plt.tight_layout()
```

#### 他の書き方

```python
# 4x4でサイズ指定された空のfigureを作成
fig=plt.figure(figsize=(4,4)) 

# figの中に、1行1列のaxesをfigの中に作成
ax=fig.add_subplot(1,1,1)

# axの中に実際にグラフを描画
ax.plot(x,y,label="y=x^2")
ax.plot(x,z,label="y=(x-1)^3")
ax.plot(x,w,label="y=((x-2)(x-4))^2")


# axesに対して設定する場合はset_が先頭につくことに注意
ax.legend()
```

```python
# 1行3列のaxesを生成
fig,ax=plt.subplots(1,3,figsize=(14,4))
print(ax)#axはarray形式であることを確認

ax[0].plot(x,y,label="y=x^2")
ax[1].plot(x,z,label="y=(x-1)^3")
ax[2].plot(x,w,label="y=((x-2)(x-4))^2")

# for文で三つのグラフを順番に描画
for i in range(3):
    ax[i].legend()
    ax[i].set_xlabel("x axis")
    ax[i].set_ylabel("y axis")
plt.tight_layout()
```

```python
# x=np.linspace(-3,10,20)
# 2行2列のaxesを生成
fig,ax=plt.subplots(2,2,figsize = (14,6))

# np.array形式であることを利用して描画
ax[0,0].plot(x,y,label="y=x^2")
ax[0,0].legend()

ax[0,1].plot(x,z,label="y=(x-1)^3")
ax[0,1].legend()

ax[1,0].plot(x,w,label="y=((x-2)(x-4))^2")
ax[1,0].legend()

#ax[1,1]に何か描いてみましょう
```

# Matplotlibの基本事項 演習

```python
# 3行1列のaxesを生成
fig,ax=plt.subplots(3,1,figsize = (8,6))

x = np.linspace(-1*np.pi, np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
ax[0].plot(x,y1,label="y=sin(x)", color = "red")
ax[1].plot(x,y2,label="y=cos(x)", color = "blue")
ax[2].plot(x,y3,label="y=tan(x)", color  = "green")

# グラフ共通の装飾
for i in range(3):
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
    ax[i].legend()

plt.tight_layout()
```

```python
#あるデータ列の各要素の数を棒グラフで知りたい
# 各棒のxの値
left=range(4)
# 曜日を取得
day=tips.day.value_counts().keys() #Index(['Sat', 'Sun', 'Thur', 'Fri'], dtype='object')
# 棒の高さ
height=tips.day.value_counts().values #array([87, 76, 62, 19])
# 棒グラフの表示
# leftで与えられたxの値を、tick_labelに代入されたdayの文字列で置き換えている
plt.bar(left,height,tick_label=day)
```

```python
day=tips.day.value_counts()
#day=tips["day"].value_counts()
day_key=tips["day"].value_counts().keys()
day_v=tips["day"].value_counts().values
print(day)
print(day_key)
print(day_v)
```

```python
# あるデータ列の各要素の全体に対する割合が知りたい
# 円グラフの表示
# autopctはフォーマット指定子、%.3f%%とすれば、小数第３位まで表示できる
```

```python
plt.pie(tips.day.value_counts(),autopct="%.1f%%",labels=tips.day.value_counts().keys())
```

```python
tips
```

```python
# あるデータ列(total_bill)の値の分布の様子が知りたい
# ヒストグラムの表示 1変数用
# bins: 縦棒の数 color: 色の指定
plt.hist(tips.total_bill,bins=20,color="green")
plt.xlabel("total_bill")
plt.ylabel("counts")
plt.title("histgram of total_bill")
```

```python
# 2つのデータ列の値の分布の様子が知りたい
# 散布図の表示
# 引数: (x座標, y座標) = (tips["total_bill"], tips["tip"])
plt.scatter(x=tips["total_bill"],y=tips["tip"])
plt.xlabel("total_bill")
plt.ylabel("tip")
```

```python
plt.scatter("total_bill","tip",data=tips) #自動で対応してくれる
plt.xlabel("total_bill")
plt.ylabel("tip")
```
